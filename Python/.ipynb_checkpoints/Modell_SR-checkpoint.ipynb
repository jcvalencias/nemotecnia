```python
from gtts import gTTS

import os

import librosa
import soundfile as sf
from pathlib import Path
SAMPLE_RATE = 22050

def doFileStuff(line, isSlow, lang, path = 'placeholder'):
    myobj = gTTS(text=line, lang=lang, slow=isSlow)
    myobj.save(path + ".mp3")
    
    y, sr = librosa.load(os.fspath("placeholder.mp3"))
    data = librosa.resample(y, sr, SAMPLE_RATE)
    sf.write('./placeholder.wav', y, sr)
    return data
```

```python
data['relacion1'] = ''
data['relacion2'] = ''

for i in range(len(data)):
    mytext = " ".join(data.iloc[i])

    doFileStuff(mytext, False, 'en')

    audio_source = sr.WavFile('placeholder.wav')

    with audio_source as source:
        audio = r.record(source)

    text = r.recognize_sphinx(audio, language="es-es")

    data.iloc[i, 2] = text

    print(text)
    
    doFileStuff(mytext, False, 'es')

    audio_source = sr.WavFile('placeholder.wav')

    with audio_source as source:
        audio = r.record(source)

    text = r.recognize_sphinx(audio, language="es-es")

    data.iloc[i, 3] = text

    print(text)
```

```python
data['relacion3'] = ''
data['relacion4'] = ''

for i in range(len(data)):
    mytext = data['Ingles'][i]

    doFileStuff(mytext, False, 'en')

    audio_source = sr.WavFile('placeholder.wav')

    with audio_source as source:
        audio = r.record(source)

    text = r.recognize_sphinx(audio, language="es-es")

    data.iloc[i, 4] = text

    print(text)
    
    doFileStuff(mytext, False, 'es')

    audio_source = sr.WavFile('placeholder.wav')

    with audio_source as source:
        audio = r.record(source)

    text = r.recognize_sphinx(audio, language="es-es")

    data.iloc[i, 5] = text

    print(text)
```
